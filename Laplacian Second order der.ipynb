```python
# Laplacian 2nd order derivative
import numpy as np
import cv2
import matplotlib.pyplot as plt
from PIL import Image
# img = cv2.imread('CAT.jpg')
arr=np.load('D:\\numpy_files\\ABS\\ABS_M01_RF000.npy')
img=arr[0:,:]
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
blur_img = cv2.GaussianBlur(img, (13, 13), 0)
edges1 = cv2.Laplacian(img,-1,  ksize=5, scale=1, delta=0, borderType=cv2.BORDER_DEFAULT)
edges = cv2.Laplacian(blur_img, -1, ksize=5, scale=1, delta=0, borderType=cv2.BORDER_DEFAULT)
output = [img, edges1]
titles = ['Original','Without Blur', 'Passed through HPF']
for i in range(3):
    plt.subplot(1, 3, i + 1)
    plt.imshow(output[i], cmap='gray')
    plt.title(titles[i])
    plt.xticks([])
    plt.yticks([])
plt.show()
# cv2.imshow("image", image.astype(np.uint8))
```
