```python
from matplotlib import pyplot as plt
from scipy import stats
import numpy as np  
import os
import pandas as pd
import csv
data = "D:/new_np/PC/"
x=[]
y=[]


for im in os.listdir(os.path.join(data)):
    phase_angle=im.split(".")[0][-3:]
    load= np.load(data+im)
    image =load[1,:,:]
    IQR = stats.iqr(image, interpolation = 'midpoint')
    x.append(str(phase_angle))
    y.append(str(IQR))
    
    #print(str(phase_angle)+"="+str(IQR))
print('x=',x)
print('y=',y)
plt.xlabel("phase_angle")
plt.ylabel("IQR")
plt.show(x,y)
x0="phase angle"
y0="iQR"
data1=pd.DataFrame({x0:x,y0:y})
data1.to_csv("IQR_PC.csv",index=False,header=True)
print('process complete')
```

```python
c=[]
d=[]
with open('IQR_PC.csv') as excelfile:
    plots = csv.reader(excelfile)
    for row in plots:
        c.append(row[0])
        d.append((row[1]))
plt.scatter(c,d,color='g')
plt.legend()
plt.show()
```
